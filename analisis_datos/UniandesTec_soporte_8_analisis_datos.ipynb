```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("/Users/cristianrinconbaron/Desktop/IntentoAnalitica/productividad_limpio.csv")
display(data)
```

```python
data.head()
data = data.drop('date', axis=1)
data
```

```python
data_parcial = data.iloc[:,0:14]
display(data_parcial)
```

```python
# Separar las variables numÃ©ricas y categÃ³ricas
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = data.select_dtypes(include=['object', 'category']).columns


data[numerical_columns].hist(figsize=(15, 12), bins=20, color='steelblue', edgecolor='black')
plt.suptitle('Histogramas de Variables NumÃ©ricas', fontsize=16)


for ax in plt.gcf().get_axes():  
    ax.grid(False)

plt.tight_layout(rect=[0, 0, 1, 0.96]) 
plt.show()


for col in categorical_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=data[col], palette='viridis')  
    plt.title(f'Histograma de {col}', fontsize=14)
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45) 
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar todas las columnas numÃ©ricas del DataFrame
numerical_data = data_parcial.select_dtypes(include=['int64', 'float64'])

# Calcular la matriz de correlaciÃ³n
corr = numerical_data.corr()

correlaciones_actual_prod = corr['actual_productivity'].abs().sort_values(ascending=False)


top_variables = correlaciones_actual_prod.index[1:4]  # Seleccionamos las 3 mÃ¡s correlacionadas


for var in top_variables:
    plt.figure(figsize=(6, 4))
    sns.regplot(x=numerical_data[var], y=numerical_data['actual_productivity'], line_kws={"color": "blue"})
    plt.title(f'DispersiÃ³n entre actual_productivity y {var}', fontsize=14)
    plt.xlabel(var)
    plt.ylabel('actual_productivity')
    plt.tight_layout()
    plt.show()
```
