# Analisis de datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("../data/productividad_limpio.csv")
display(data)
```

```python
data.head()
data = data.drop('date', axis=1)
data
```

```python
data_parcial = data.iloc[:,0:14]
display(data_parcial)
```

```python
# Separar las variables numéricas y categóricas
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = data.select_dtypes(include=['object', 'category']).columns


data[numerical_columns].hist(figsize=(15, 12), bins=20, color='steelblue', edgecolor='black')
plt.suptitle('Histogramas de Variables Numéricas', fontsize=16)


for ax in plt.gcf().get_axes():  
    ax.grid(False)

plt.tight_layout(rect=[0, 0, 1, 0.96]) 
plt.show()


for col in categorical_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=data[col], palette='viridis')  
    plt.title(f'Histograma de {col}', fontsize=14)
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45) 
    plt.tight_layout()
    plt.show()
```

```python
numerical_data = data_parcial.select_dtypes(include=['int64', 'float64'])
corr = numerical_data.corr()

#  heatmap 
plt.figure(figsize=(12, 8))  
sns.heatmap(corr, cmap="Blues", annot=True, fmt=".2f", linewidths=0.5, square=True, 
            cbar_kws={"shrink": .8}, vmin=-1, vmax=1)

# Ajustar las etiquetas
plt.xticks(rotation=45, ha="right", fontsize=10)  
plt.yticks(rotation=0, fontsize=10)  
plt.title("Matriz de Correlación", fontsize=16)

plt.tight_layout()  
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar todas las columnas numéricas del DataFrame
numerical_data = data_parcial.select_dtypes(include=['int64', 'float64'])

# Calcular la matriz de correlación
corr = numerical_data.corr()

# Identificar la variable con mayor correlación con actual_productivity (sin contar a sí misma)
correlaciones = corr['actual_productivity'].drop('actual_productivity')
max_corr_var = correlaciones.idxmax()  # Variable con la correlación más alta

# Heatmap 
plt.subplot(1, 2, 2)
sns.heatmap(correlaciones.to_frame(), annot=True, cmap="Blues", cbar=True, vmin=-1, vmax=1)
plt.title('Correlaciones de la variable de interés')
plt.tight_layout()

plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Separar las variables numéricas y categóricas
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = data.select_dtypes(include=['object', 'category']).columns

# Crear gráficos de violín para las variables numéricas
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns) // 3 + 1, 3, i+1)
    sns.violinplot(x=data[col], palette='Blues', inner='quartile')
    plt.title(f'Gráfico de Violín: {col}', fontsize=14)
    plt.xlabel(col)
    plt.ylabel('Densidad')

plt.suptitle('Gráficos de Violín de Variables Numéricas', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Crear gráficos de barras para las variables categóricas
for col in categorical_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=data[col], palette='viridis')
    plt.title(f'Gráfico de Barras: {col}', fontsize=14)
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    

```
